### Step 5. Delete the column 'Unnamed: 0' and 'Id'


```python
# drop function is really useful
# df.drop('col_name',1) 1 for col 0 for row

# baby_names.drop('Id', 1)
baby_names = baby_names.drop(['Unnamed: 0', 'Id'], 1) # I used [] to include the 2 cols that I want to delet
```

### Step 6. Is there more male or female names in the dataset?

```python
# this method not correct!
# this method calculate the number of male and female in the Gender col, however, we need to know the number with respect 
# to the Count col

male = []
female = []
empty = []
for mn in baby_names['Gender']:
    if mn == 'M':
        male.append(mn)
    elif mn == 'F':
        female.append(mn)
    else:
        empty.append(mn)
        
        
len(male), len(female), len(empty)        
```

```python
# Correct way

# first we grouped the Gender col by F and M, then we sum the numbers form the Count col
N = baby_names.groupby('Gender').agg({
    'Count': 'sum'
})
N
# number / total
N / baby_names.Count.sum()
```

### Step 7. Group the dataset by name and assign to names


```python
names = baby_names.groupby('Name').agg({
    'Count': 'sum'
})
names
```

### Step 8. How many different names exist in the dataset?

### Step 9. What is the name with most occurrences?


```python
no = names/len(names)
max_x = no.loc[no["Count"].idxmax()]
max_x
```
